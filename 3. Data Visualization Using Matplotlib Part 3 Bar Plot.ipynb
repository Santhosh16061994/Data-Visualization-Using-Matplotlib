```python
''' 
we are going to solve the following questions to understand the barplot.
Samples:
---------------------------------------------------------------

Grades:  A    B   C    D
Test1:   3    13  10   4
Test2:   5    14  6    5
Test3:   7    15  5    3
------------------------------------------------------------------

grades = ['A','B','C','D']
test1=[3,13,10,4]
test2=[5,14,6,5]
test3=[7,15,5,3]
'''
import pandas as pd
```

```python
grades = ['A','B','C','D']
test1=[3,13,10,4]
test2=[5,14,6,5]
test3=[7,15,5,3]
```

# draw a bar plot

```python
plt.bar(grades,test1)
plt.show()
```

# give x label y label and title

```python
plt.bar(grades,test1)


plt.title('Test 1 Grades ') # title
plt.xlabel('Grades')
plt.ylabel('No.of Students') # we are giving by our own
```

# fornt size of title , x and y label

```python
plt.bar(grades,test1)


plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
# we can change font size
```

```python
plt.bar(grades,test1,width=0.3,align='edge') # editing width and align
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()
```

# colour of bar and edge colour

```python
plt.bar(grades,test1,width=0.3,align='edge',color='y') # colour
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()
```

```python
plt.bar(grades,test1,width=0.3,align='edge',color=['y','k','b','g',]) # different colour in list
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()
```

```python
plt.bar(grades,test1,width=0.3,align='edge',color='y',edgecolor='black') # colour of edge
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()
```

# change line style and line width

```python
plt.bar(grades,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':') 
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()
```

```python
plt.bar(grades,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',linewidth=4) # line width lw
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()
```

```python
plt.bar(grades,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4) # line width lw
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()
```

# bar plot side by side of test1 and test 2

```python
plt.bar(grades,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4) 
plt.bar(grades,test2,width=0.3,align='edge',color='b',edgecolor='black',
linestyle=':',lw=4) 
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()
```

```python
# to create side by side ceate numpy array
index=np.arange(len(grades))
index


```

```python
plt.bar(index,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4) 
plt.bar(index+0.3,test2,width=0.3,align='edge',color='b',edgecolor='black',
linestyle=':',lw=4) 
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()

# here we created numpy array
# added scaler value 
# removed grades from original codes
```

```python
plt.bar(index,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4) 
plt.bar(index+0.5,test2,width=0.3,align='edge',color='b',edgecolor='black',
linestyle=':',lw=4) 
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.show()

# changed the scalr value .3 - .5 
```

```python
# inplace of grade it showing floating point


plt.bar(index,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4) 
plt.bar(index+0.5,test2,width=0.3,align='edge',color='b',edgecolor='black',
linestyle=':',lw=4) 
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.xticks(index,grades)
plt.show()

# xtick method tpo add grades




```

```python
# inplace of grade it showing floating point


plt.bar(index,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4) 
plt.bar(index+0.3,test2,width=0.3,align='edge',color='b',edgecolor='black',
linestyle=':',lw=4) 
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.xticks(index+0.3,grades) # added scalr value t move left
plt.show()

# xtick method tpo add grades




```

```python
plt.bar(index,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4) 
plt.bar(index+0.3,test2,width=0.3,align='edge',color='b',edgecolor='black',
linestyle=':',lw=4) 
plt.bar(index+0.6,test2,width=0.3,align='edge',color='g',edgecolor='black',
linestyle=':',lw=4)
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.xticks(index+0.5,grades)
plt.show()

```

# legend

```python
plt.bar(index,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4,label='Test1') 
plt.bar(index+0.3,test2,width=0.3,align='edge',color='b',edgecolor='black',
linestyle=':',lw=4,label='Test2') 
plt.bar(index+0.6,test2,width=0.3,align='edge',color='g',edgecolor='black',
linestyle=':',lw=4,label='Test3') # added lable legend
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.xticks(index+0.5,grades)
plt.legend()
plt.show()

```

```python
plt.bar(index,test1,width=0.3,align='edge',color='y',edgecolor='black',
linestyle=':',lw=4,label='Test1') 
plt.bar(index+0.3,test2,width=0.3,align='edge',color='b',edgecolor='black',
linestyle=':',lw=4,label='Test2') 
plt.bar(index+0.6,test2,width=0.3,align='edge',color='g',edgecolor='black',
linestyle=':',lw=4,label='Test3')
plt.title('Test 1 Grades ',fontsize=15) # title
plt.xlabel('Grades',fontsize=15)
plt.ylabel('No.of Students',fontsize=15) # we are giving by our own
plt.xticks(index+0.5,grades,rotation=90) # x axis label rotation when larger in size
plt.legend()
plt.show()

```
