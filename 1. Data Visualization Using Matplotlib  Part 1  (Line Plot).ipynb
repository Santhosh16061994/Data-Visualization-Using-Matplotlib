# draw a simple line plot

It displays a sequence of data points as markers. The points are ordered typically by their x-axis value. These points are joined with straight line segments. A line graph is used to visualize a trend in data over intervals of time.21-Dec-2021

It shows related information by drawing a continuous line between all the points on a grid. Line charts compare two variables: one is plotted along the x-axis (horizontal) and the other along the y-axis (vertical).02-Sept-2021

Why line plot is used?
Image result for why line plot used in data science
A line graph—also known as a line plot or a line chart—is a graph that uses lines to connect individual data points. A line graph displays quantitative values over a specified time interval. In finance, line graphs are commonly used to depict the historical price action of an asset or security.

A line chart (aka line plot, line graph) uses points connected by line segments from left to right to demonstrate changes in value. The horizontal axis depicts a continuous progression, often that of time, while the vertical axis reports values for a metric of interest across that progression.

# font size of title , x lable , y lable and also change location

```python
x=[1,2,3,7]
y= [1,7,9,8]
plt.plot(x,y)
plt.title('Line Plot Demo',fontsize=20,loc='left')
plt.xlabel('JIRA',fontsize=20,loc='right')
plt.ylabel('WGS',fontsize=20,loc='bottom')

plt.show()
```

# colour and width of the line

```python
x=[1,2,3,7]
y= [1,7,9,8]
plt.plot(x,y,color='r',linewidth=4)
plt.title('Line Plot Demo',fontsize=20,loc='left')
plt.xlabel('JIRA',fontsize=20,loc='right')
plt.ylabel('WGS',fontsize=20,loc='bottom')

plt.show()
```

# Adding legend to the graph and its location

```python
x=np.random.randint(low=1,high=10,size=20)
# we are generating random no. from 1-9 1st inclusive and last exclusive such 20 numbers
y=np.random.randint(low=1,high=8,size=20)
print(x)
print(y)
plt.plot(x,y,'r--D',linewidth=4,markersize=10,markerfacecolor='blue',markeredgecolor='blue',
linestyle=':')
plt.title('Line Plot Demo',fontsize=20,loc='left')
plt.xlabel('JIRA',fontsize=20,loc='right')
plt.ylabel('WGS',fontsize=20,loc='bottom')
plt.legend(["A Sample Line Plot"],loc=0)

plt.show()
```

# value of x lim an y lim

```python
x=np.random.randint(low=1,high=10,size=20)
# we are generating random no. from 1-9 1st inclusive and last exclusive such 20 numbers
y=np.random.randint(low=1,high=8,size=20)

plt.plot(x,'r--D',linewidth=4,markersize=5,markerfacecolor='blue',markeredgecolor='blue',
linestyle=':',label="Line 1")
plt.plot(y,'b--D',linewidth=4,markersize=5,markerfacecolor='blue',markeredgecolor='blue',
linestyle=':',label="Line 2")
plt.title('Line Plot Demo',fontsize=20,loc='left')
plt.xlabel('JIRA',fontsize=20,loc='right')
plt.ylabel('WGS',fontsize=20,loc='bottom')
plt.legend()
plt.xlim(0,20)
plt.ylim(0,15)
# we can costomise value of x axis and y axis

plt.show()

```

# Chane the Figure Size

```python
x=np.random.randint(low=1,high=10,size=20)
# we are generating random no. from 1-9 1st inclusive and last exclusive such 20 numbers
y=np.random.randint(low=1,high=8,size=20)

plt.figure(figsize=(10,3)) # changin the fig size
plt.plot(x,'r--D',linewidth=4,markersize=5,markerfacecolor='blue',markeredgecolor='blue',
linestyle=':',label="Line 1")
plt.plot(y,'b--D',linewidth=4,markersize=5,markerfacecolor='blue',markeredgecolor='blue',
linestyle=':',label="Line 2")
plt.title('Line Plot Demo',fontsize=20,loc='left')
plt.xlabel('JIRA',fontsize=20,loc='right')
plt.ylabel('WGS',fontsize=20,loc='bottom')
plt.legend()
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.grid(color='b',linestyle='--',linewidth=2) # Grid added to the plot
plt.show()

```
