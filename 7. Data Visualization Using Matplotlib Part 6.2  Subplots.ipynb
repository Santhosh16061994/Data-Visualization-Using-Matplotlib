# 4plots

```python
# here we are accessing figure  axex level object
fig,((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,ncols=2) # first wecode  fig lev ax
fig.suptitle('Iris_dataset')

ax1.hist(data.SepalLengthCm,color='g')
ax1.set_title('SepalLengthCm') # for giving titel of one graph
ax1.set_xlabel('SepalLengthCm')
ax1.set_ylabel('Counts')

ax2.hist(data.PetalLengthCm,color='k')
ax2.set_title('PetalLengthCm') # for giving titel of one graph
ax2.set_xlabel('PetalLengthCm')
ax2.set_ylabel('Counts')

ax3.hist(data.SepalWidthCm,color='r')
ax3.set_title('SepalWidthCm') # for giving titel of one graph
ax3.set_xlabel('SepalWidthCm')
ax3.set_ylabel('Counts')

ax4.hist(data.PetalWidthCm,color='y')
ax4.set_title('PetalWidthCm') # for giving titel of one graph
ax4.set_xlabel('PetalWidthCm')
ax4.set_ylabel('Counts')

plt.tight_layout() # -- remove overlapping of graph or count label
plt.show()
```

# share x and share y

```python
# here we are accessing figure  axex level object
fig,((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,ncols=2,sharex=True,sharey=True) # first wecode  fig lev ax
fig.suptitle('Iris_dataset')

ax1.hist(data.SepalLengthCm,color='g')
ax1.set_title('SepalLengthCm') # for giving titel of one graph
ax1.set_xlabel('SepalLengthCm')
ax1.set_ylabel('Counts')

ax2.hist(data.PetalLengthCm,color='k')
ax2.set_title('PetalLengthCm') # for giving titel of one graph
ax2.set_xlabel('PetalLengthCm')
ax2.set_ylabel('Counts')

ax3.hist(data.SepalWidthCm,color='r')
ax3.set_title('SepalWidthCm') # for giving titel of one graph
ax3.set_xlabel('SepalWidthCm')
ax3.set_ylabel('Counts')

ax4.hist(data.PetalWidthCm,color='y')
ax4.set_title('PetalWidthCm') # for giving titel of one graph
ax4.set_xlabel('PetalWidthCm')
ax4.set_ylabel('Counts')

plt.tight_layout() # -- remove overlapping of graph or count label
plt.show()
```

# change fig size

```python
# here we are accessing figure  axex level object

fig,((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,ncols=2,sharex=True,sharey=True,figsize=(10,6))
fig.set_edgecolor('k')
fig.set_linewidth(6) # chane fig size
fig.set_facecolor('lightgreen')
fig.suptitle('Iris_dataset')

ax1.hist(data.SepalLengthCm,color='g')
ax1.set_title('SepalLengthCm') # for giving titel of one graph
ax1.set_xlabel('SepalLengthCm')
ax1.set_ylabel('Counts')

ax2.hist(data.PetalLengthCm,color='k')
ax2.set_title('PetalLengthCm') # for giving titel of one graph
ax2.set_xlabel('PetalLengthCm')
ax2.set_ylabel('Counts')

ax3.hist(data.SepalWidthCm,color='r')
ax3.set_title('SepalWidthCm') # for giving titel of one graph
ax3.set_xlabel('SepalWidthCm')
ax3.set_ylabel('Counts')

ax4.hist(data.PetalWidthCm,color='y')
ax4.set_title('PetalWidthCm') # for giving titel of one graph
ax4.set_xlabel('PetalWidthCm')
ax4.set_ylabel('Counts')

plt.tight_layout() # -- remove overlapping of graph or count label
plt.show()
```
