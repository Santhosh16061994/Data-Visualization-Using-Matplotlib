```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  
```

# survey

```python

movie = ['Comedy','Action','Romance','Drama']
votes=[5,6,3,1]
```

1. Draw A Pie Chart


```python
plt.pie(votes)
plt.show
```

2. Give Title, Label (Also Change Label Distance), And Legend 


```python
plt.title('Favorite Type of Moive ')

plt.pie(votes,labels=movie)
plt.show
```

```python
#legend
plt.title('Favorite Type of Moive ')
plt.pie(votes,labels=movie,labeldistance=1.9) # labeldistance
plt.legend(loc=4)
plt.show
```

3. Pull out Action and Drama Wedges


```python
#legend
plt.title('Favorite Type of Moive ')
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,]) 
plt.legend(loc=4)
plt.show()

# pullin oot wedges
# 4 variable so given 4 '   '
```

4. Change The Start Angle To 90 Degree


```python
plt.title('Favorite Type of Moive ')
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90)  # changing start angle
plt.legend(loc=4)
plt.show()
```

5. Set The Color of Each Wedge


```python
plt.title('Favorite Type of Moive ')
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'])  # chaing colour in wedge
plt.legend(loc=4)
plt.show()
```

6. Display Percentage & Change Distance Inside Each Wedge


```python
plt.title('Favorite Type of Moive ')
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'],
autopct='%0.2f%%')   # % diplay on wedge
plt.legend(loc=4)
plt.show()
```

```python
plt.title('Favorite Type of Moive ')
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'],
autopct='%0.2f%%')   # % diplay on wedge
plt.legend(loc=4)
plt.show()
plt.title('Favorite Type of Moive ')
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'],
autopct='%0.2f%%',
pctdistance=0.8)   # % diplay on wedge will move sides
plt.legend(loc=4)
plt.show()
```

#7. Change Radius of The Pie Chart


```python

plt.title('Favorite Type of Moive ')
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'],
autopct='%0.2f%%',
pctdistance=0.8,
radius=2)   # chnge the size
plt.legend(loc=4)
plt.show()
```

```python

plt.title('Favorite Type of Moive ',y=1.5) # title will change
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'],
autopct='%0.2f%%',
pctdistance=0.8,
radius=2)   # chnge the size
plt.legend(loc=4)
plt.show()
```

# 8. Set Labels Size of The Pie Chart


```python

plt.title('Favorite Type of Moive ',y=1.5) # title will change
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'],
autopct='%0.2f%%',
pctdistance=0.8,
radius=1,
textprops={'fontsize':12})   # chnge the label size
plt.legend(loc=4)
plt.show()
```

9. Set Border of The Wedges


```python

plt.title('Favorite Type of Moive ',y=1.5) # title will change
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'],
autopct='%0.2f%%',
pctdistance=0.8,
radius=1,
textprops={'fontsize':12},
wedgeprops={'edgecolor':'black','linewidth':2,'ls':':'})   # # border of wedge by this parameter
plt.legend(loc=4)
plt.show()
```

10. Add A shadow to the Pie Chart


```python

plt.title('Favorite Type of Moive ',y=1.5) # title will change
plt.pie(votes,labels=movie,labeldistance=1.1,
explode=[0,0.1,0,0.1,],
startangle=90,
colors=['r','y','b','m'],
autopct='%0.2f%%',
pctdistance=0.8,
radius=1,
textprops={'fontsize':12},
wedgeprops={'edgecolor':'black','linewidth':2,'ls':':'},
shadow=True)   #adding shadow to the chart
plt.legend(loc=4)
plt.show()
```
