```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  
```

sctter plot is used to observe relationship b/w 2 numeric variable

temp= [14.2,16.4,11.9,15.2,18.5,22.1,19.4]

```python
temp= [14.2,16.4,11.9,15.2,18.5,22.1,19.4]
icecream_sales=[215,325,185,332,406,522,412]
tea_sales=[80,65,89,70,56,45,43]
```

1. Draw A Scatter Plot With Title, Xlabel, And YLabel


```python
plt.scatter(temp,icecream_sales)
plt.title('Temp VS Icecream_sales')#3 added title
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

2. Change Size of The-Dots


```python
plt.scatter(temp,icecream_sales,s=100) # size of dot
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

3. Change Color of The-Dots


```python
plt.scatter(temp,icecream_sales,s=100,c='red') # colour of data points
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

```python
plt.scatter(temp,icecream_sales,s=100,c='red',marker='v') # shape of data points
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

```python
plt.scatter(temp,icecream_sales,s=100,c='red',marker=7) # shape of data points
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

5. Change The Edge Color of The Marker Also change Line Width


```python
plt.scatter(temp,icecream_sales,s=100,c='red',marker='o',ec='k',lw=2) # marker edge and line width ec=edgecolor , lw linewidth
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

6. Change Alpha Parameter


```python
plt.scatter(temp,icecream_sales,s=100,c='red',marker='o',ec='k',lw=2,alpha=0.25) 
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

```python
plt.scatter(temp,icecream_sales,s=100,c='red',marker='o',ec='k',lw=2,alpha=0.75)  # alpha is 0 .25 - 1
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

7. Set Different Colors For Different Dots


```python
plt.scatter(temp,icecream_sales,s=100,c=[11, 12, 13, 14, 15, 16, 17],marker='o',ec='k',lw=2,alpha=0.75)  # in red we are pasing color list
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

```python
plt.scatter(temp,icecream_sales,s=100,c=color,marker='o',ec='k',lw=2,alpha=0.75)  # in red we are pasing color list
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

# 8. Use In-built Color Maps


```python
plt.scatter(temp,icecream_sales,s=100,c=color,marker='o',ec='k',lw=2,alpha=0.75,
cmap='Oranges')  # in builyt color map in link https://matplotlib.org/stable/tutorials/colors/colormaps.html
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

9. Add Colorbar 



```python
plt.scatter(temp,icecream_sales,s=100,c=color,marker='o',ec='k',lw=2,alpha=0.75,
cmap='Oranges')  
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.colorbar() # colour bar added
plt.show()
```

# 10. Customize The Limit of Colorbar 


```python
plt.scatter(temp,icecream_sales,s=100,c=color,marker='o',ec='k',lw=2,alpha=0.75,
cmap='Oranges',
vmin=10,vmax=20)  # coustomise the colour bar
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.colorbar()
plt.show()
```

# 10. label the name of  Colorbar 


```python
plt.scatter(temp,icecream_sales,s=100,c=color,marker='o',ec='k',lw=2,alpha=0.75,
cmap='Oranges',
vmin=10,vmax=20)  # coustomise the colour bar
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
c_bar = plt.colorbar()
c_bar.set_label('Temp VS Icecream_sales')
plt.show()
```

# 12. Draw Two Scatter Plots In One Figure


```python
plt.scatter(temp,icecream_sales,s=100,c='green',marker='o',ec='k',lw=2,alpha=1,label='icecream_sales')
plt.scatter(temp,tea_sales,s=100,c='yellow',marker='o',ec='k',lw=2,alpha=1,label='tea_sales')
plt.legend()
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```

# 14. Change Figure Size & Color


```python
plt.figure(figsize=(10,3),facecolor='lightgreen',edgecolor='black',linewidth=5)  
plt.scatter(temp,icecream_sales,s=100,c='green',marker='o',ec='k',lw=2,alpha=1,label='icecream_sales')
plt.scatter(temp,tea_sales,s=100,c='yellow',marker='o',ec='k',lw=2,alpha=1,label='tea_sales')
plt.legend()
plt.title('Temp VS Icecream_sales')
plt.xlabel('Temp')
plt.ylabel('icecream_sales')
plt.show()
```
